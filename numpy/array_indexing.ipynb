```python
import numpy as np
```

## Numpy Indexing and Slicing of Vectors

```python
arr = np.arange(11,21)
arr
```

```python
arr_slice = arr[3::2]
arr_slice
```

```python
arr[4]
```

## Numpy Indexing and Slicing of Matrix

```python
arr = np.arange(1,31).reshape(6,5)
arr
```

```python
slice = arr[3:,3:]
slice
```

```python
arr[:,2]
```

```python
arr[4,:]
```

## Boolean Indexing

```python
arr = np.arange(11,21)
arr
```

```python
bool_index = arr % 2 == 0
bool_index
```

```python
arr = arr[bool_index]
arr
```
