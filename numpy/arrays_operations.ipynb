### Advance Array Manipulation

#### Stacking Array

```python
a = np.array([1, 2, 3, 4])
b = np.array([5, 6, 7, 8])
```

```python
np.vstack((a,b))
```

```python
np.hstack((a,b))
```

```python
np.column_stack((a,b))
```

#### Splitting Array

```python
c = np.arange(16).reshape(4, 4)
c
```

```python
np.hsplit(c,4)
```

```python
np.vsplit(c,2)
```
