### Valid Sudoku

```python
import numpy as np
```

```python
s = np.array([
    [5, 3, 4, 6, 7, 8, 9, 1, 2],
    [6, 7, 2, 1, 9, 5, 3, 4, 8],
    [1, 9, 8, 3, 4, 2, 5, 6, 7],
    
    [8, 5, 9, 7, 6, 1, 4, 2, 3],
    [4, 2, 6, 8, 5, 3, 7, 9, 1],
    [7, 1, 3, 9, 2, 4, 8, 5, 6],
    
    [9, 6, 1, 5, 3, 7, 2, 8, 4],
    [2, 8, 7, 4, 1, 9, 6, 3, 5],
    [3, 4, 5, 2, 8, 6, 1, 7, 9]
])
```

```python
l = np.sum(s, axis=0) 
for i in l:
    if i != 45:
        print('for rows, sudoku is not valid!')
        break
else:
    print('for rows, sudoku is valid!')
```

```python
l = np.sum(s, axis=1) 
for i in l:
    if i != 45:
        print('for columns, sudoku is not valid!')
        break
else:
    print('for columns, sudoku is valid!')
```

```python
# Check 3x3 subgrids
valid = True

for i in range(0, 9, 3):        # row blocks
    for j in range(0, 9, 3):    # column blocks
        subgrid = s[i:i+3, j:j+3]
        print(subgrid)
        if np.sum(subgrid) != 45:
            print("3x3 grid is not valid!")
            valid = False
            break
    if not valid:
        break

if valid:
    print("3x3 grid is valid!")
```
